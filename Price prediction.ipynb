```python
x_train.dropna(subset=['MSZoning', "Exterior1st", "Exterior2nd", "BsmtFinSF1",], inplace=True)
```

```python
x_train['Utilities'].fillna("AllPub", inplace=True)
x_train['MasVnrType'].fillna("BrkFace", inplace=True)
x_train['LotFrontage'].fillna(round(x_train['LotFrontage'].mean()), inplace=True)
x_train['MasVnrArea'].fillna(0.0, inplace=True)
x_train['BsmtQual'].fillna('TA', inplace=True)
x_train['BsmtCond'].fillna('TA', inplace=True)
x_train['BsmtExposure'].fillna('No', inplace=True)
x_train['BsmtFinType1'].fillna('GLQ', inplace=True)
x_train['BsmtFinType2'].fillna('Unf', inplace=True)
x_train['BsmtFullBath'].fillna(0.0, inplace=True)
x_train['BsmtHalfBath'].fillna(0.0, inplace=True)
x_train['KitchenQual'].fillna('TA', inplace=True)
x_train['Functional'].fillna('Typ', inplace=True)
x_train['FireplaceQu'].fillna('Gd', inplace=True)
x_train['GarageType'].fillna('Attchd', inplace=True)
x_train['GarageYrBlt'].fillna(round (x_train['GarageYrBlt'].mean()), inplace=True)
x_train['GarageFinish'].fillna('Unf', inplace=True)
x_train['GarageCars'].fillna(2.0, inplace=True)
x_train['GarageQual'].fillna('TA', inplace=True)
x_train['GarageCond'].fillna('TA', inplace=True)
x_train['SaleType'].fillna('WD', inplace=True)
x_train['Electrical'].fillna('SBrkr', inplace=True)
```

```python
x_ = x_train[x_train.Type==1]
x_ = x_.drop(["Type"], axis=1)

test = x_train[x_train.Type==0]
test = test.drop(["Type"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train_scaled_80, x_test_scaled_20, y_train_scaled_80, y_test_scaled_20 = train_test_split(x_scaled, y_, test_size=0.20, random_state=1)

```

```python
y_predict_best = model_randomforest_train80.predict(test)
y_predict_best
```
